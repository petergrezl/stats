```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(-5,5,1000)
y = norm.pdf(x, loc=0, scale=1)

points = np.logical_or((x < -1.645), (x > 1.645))
```

```python
plt.plot(x,y)
plt.fill_between(x, 0, y, facecolor='RED', where=points)
plt.xlim(-5, 5)
plt.ylim(-0, 0.45)
plt.show()
```
